# Exploratory Data Analysis 

```python
ax= sns.countplot(x='Gender',data=df)
sns.set(rc={'figure.figsize':(5,7)})
```

```python
ax= sns.countplot(x='Gender',data=df)
for bars in ax.containers:  #this is for lable in bar
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index= False) ['Amount'].sum().sort_values(by='Amount',ascending = False)
```

```python
#data sortby gender basis, and sum of amount
sale_gen=df.groupby(['Gender'],as_index= False) ['Amount'].sum().sort_values(by='Amount',ascending = False) 
sns.barplot( x='Gender',y= 'Amount', hue= 'Gender',data=sale_gen)
```

###### most of buyer are  females  

```python
df.head(2)
```

```python
ax= sns.countplot(x='Age Group',hue='Gender',data=df)
```

```python
ax= sns.countplot(x='Age Group',hue='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
    
```

###### most of the buyer are age group is 26-35 (Females)

```python
df.head(2)
```

```python
#total number of order of top10 States
total_states = df.groupby(['State'], as_index= False) ['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=total_states ,x= 'State',y= 'Orders')
```

```python
#total amount/sale fromtop 10 state
total_states = df.groupby(['State'], as_index= False) ['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=total_states ,x= 'State',y= 'Amount')
```

#### From above Graph most of the Orders are in Uttar Pradesh & most of the3 amount also from Utter Pradesh 

```python
 ax= sns.countplot(x= 'Marital_Status',hue='Gender', data= df)
sns.set(rc = {'figure.figsize': (7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

###### mostof the buyer are married woman

```python
sale_State = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending = False)
sns.barplot(x= 'Marital_Status',y= 'Amount',data=sale_State ,hue='Gender' )
sns.set(rc = {'figure.figsize': (7,5)})
```

##### most of the buyers are married females

```python
ax= sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
  
```

###### most of the peoples occupation in IT Sector then Healthcare then aviation ans so on....

```python
#occupation based on Amount 
total_states = df.groupby(['Occupation'], as_index= False) ['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=total_states ,x= 'Occupation',y= 'Amount')
```

###### Also on based on amount most of the peole occupation in It sector, heathcare avation and so on....

```python
ax = sns.countplot(x='Product_Category', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
  
```

```python
product_cate= df.groupby('Product_Category', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(data = product_cate ,x = 'Product_Category' ,y = 'Amount')
sns.set(rc={'figure.figsize':(30,5)})
```

###### we notice most of the sold product is food ,clothing ,electronic & gadgets
  

```python
sold_product = df.groupby('Product_ID', as_index=False)['Orders'].sum().sort_values(by='Orders',ascending= False)
sns.barplot(data = sold_product ,x = 'Product_ID' ,y = 'Orders')
sns.set(rc={'figure.figsize':(30,5)})
```
